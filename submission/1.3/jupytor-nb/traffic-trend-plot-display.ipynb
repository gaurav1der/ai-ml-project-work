### In this exercise, generate some basic charts using Pandas. 


```python
# import data from pandas
import pandas as pd
import os

#Make directory called plots
os.makedirs("plots", exist_ok=True)

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Users/gauravgoel/Documents/workspace/berkeley/ml-ai-examples/submission/1.4/datasets/Traffic.csv')

# Group by 'Date' and sum counts for each vehicle type
daily_counts = df.groupby('Day of the week')[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].sum()

# Plot
daily_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Daily Traffic Counts by Vehicle Type')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.legend(title='Vehicle Type')
plt.tight_layout()
plt.show()
# plt.savefig('daily_traffic_by_vehicle_type.pdf')
plt.close()
```
